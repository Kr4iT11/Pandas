```python
#Generating numbers starting from 2 to 6 
np.linspace(2,6)
```

```python
#Generating 1000 numbers starting from 2 to 6 
np.linspace(2,6,1000)
```

```python
#Reshaping the linspace generated 1d array 
np.linspace(2,6,10).reshape(5,2)
```

```python
#Providing endpoint parameter to include endpoint or not 
np.linspace(2,6,10,endpoint=False).reshape(5,2)
```

```python
#Providing retshape parameter to determine number of steps taken to generate the numbers
np.linspace(2,6,10,False,retstep=True)
```

```python
#just like linspace its logspace generating log values starting from 2 to ending 6 and generating 10 numbers
np.logspace(2,6,10)
```

```python
#Generating from 2 to 6 10 numbers log
np.logspace(2,6,10,base=10)
```

```python
#Shallow copy deep copy concept
x = np.array([1,2,3])
```

```python
y = x
```

```python
z = np.copy(x)
```

```python
z
```

```python
x
```

```python
y
```

```python
x[0] = 100
```

```python
x
```

```python
y
```

```python
z
```

```python
# as x and y are referring to same location so if changes in x then it will reflect in y vice versa known as shallow copy
# np.copy will copy and create a new memory location so no changes will be in z known as deep copy
```
